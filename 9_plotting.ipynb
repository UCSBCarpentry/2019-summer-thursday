```python
import pandas as pd #abbreviated new pandas library as pd

#created new variable called data to import oceania data
data = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col = 'country')

#Extract year from last 4 characters of each column name
years = data.columns.str.strip('gdpPercap_')

#Convert year values from string to integer, save results back to data variable
data.columns = years.astype(int)
print(data)

data.loc['Australia'].plot() #locate Australia row values, plot these only, not New Zealand
```

```python
#plot many sets of data together
print(data)
#select/stripping two countries' worth of data from data, our created dataset
#copyoforiginal
gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']

#plot with differently-colored markers
plt.plot(years, gdp_australia, 'b-', label = 'Australia')
plt.plot(years, gdp_nz, 'g-', label = 'New Zealand')

#create legend
plt.legend(loc = 'upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')
```
