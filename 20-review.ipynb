```python
import matplotlib.pyplot as plt
import pandas as pd
plt.close('all')
```

```python
#read in africa data
africa = pd.read_csv('data/gapminder_gdp_africa.csv', index_col='country')
#display(africa)
#Let's do some data clean up
#Extract year from last 4 characters of each column name
years = africa.columns.str.strip('gdpPercap_')

#Convert year values to integers, saving results back to dataframe
africa.columns = years.astype(int)

#plot
plt.figure()
africa.loc['Zimbabwe'].plot()
plt.xlabel('Year')
plt.ylabel('GDP per capita')
```

```python
#Select and transform data, then plot it
africa.plot()
africa.T.plot()

plt.ylabel('GDP per capita')
```

```python
#plot last three countries in data set
#plt.figure()
africa.loc[['Uganda', 'Zambia','Zimbabwe']].T.plot()
plt.xlabel('Year')
plt.ylabel('GDP per capita')

#africa.loc[['Uganda', 'Zambia','Zimbabwe'],['1952','1957', '1962']].T.plot()
africa.loc[['Uganda', 'Zambia','Zimbabwe'],[1952,1957,1962]].T.plot()
```
