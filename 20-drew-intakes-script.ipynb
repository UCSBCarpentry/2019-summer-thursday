```python
import pandas as pd #abbreviated new pandas library as pd

#created new variable called intakedata_2018_2019 to import intake data
intakedata_2018_2019 = pd.read_csv('drewsdata/intakedata20181-20193.csv')

#drop null value coulmns to avoid errors
intakedata_2018_2019.dropna(inplace = True)
```

```python
display(intakedata_2018_2019)
type(intakedata_2018_2019)
```

```python
#check data type in each column. 
#Are my numbers actually stored as integers?
intakedata_2018_2019.dtypes
```

```python
#take off year from quarter values

#create new object quarter_withoutyear to contain only quarter values
quarter = intakedata_2018_2019['Quarter']
type(quarter) # series

quarter_noyear = [x[:-5] for x in quarter]
display(quarter_noyear)
```

```python
#add quarter_noyear back into intakedata_2018_2019
intakedata_2018_2019["Quarter"] = quarter_noyear
display(intakedata_2018_2019)
type(intakedata_2018_2019) # still a dataframe, good!
```

```python
#add two new columns with zeroes
clinicians = [0]*len(intakedata_2018_2019)
intakes_per_clinicians = [0]*len(intakedata_2018_2019)

intakedata_2018_2019['Clinicians'] = clinicians
intakedata_2018_2019['Intakes Per Clinicians'] = intakes_per_clinicians

display(intakedata_2018_2019)
type(intakedata_2018_2019)
```

```python
#ask for quarter, week, and # clinicians:
input_quarter = input("Enter Quarter: Fall, Winter or Spring")
input_week = input("Enter Week")
input_clinicians = input("Enter Number of Clinicians")

print("Printing Quarter, Week, # Clinicians")
print(input_quarter, input_week, input_clinicians)
```
